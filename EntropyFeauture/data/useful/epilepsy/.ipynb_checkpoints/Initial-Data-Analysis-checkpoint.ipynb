```python
limited_control = [list(x.groupby('date'))[0:y] for x, y in zip(df_control,split_scores_control[1]['days'])]
limited_condition = [list(x.groupby('date'))[0:y] for x, y in zip(df_condition,split_scores_condition[1]['days'])]
```

```python
#untupled_list = [[x[1] for y in limited_control] for x in y]

untupled_list = []
for y in limited_control:
    untupled_list.append([x[1] for x in y])
limited_control_df = [pd.concat(x) for x in untupled_list]

untupled_list = []
for y in limited_condition:
    untupled_list.append([x[1] for x in y])
limited_condition_df = [pd.concat(x) for x in untupled_list]

```

### Extracting the following features from selected data

<ol>
<li>Mean Activity Level</li>
<li>Standard Deviation</li>
<li>Proportion of Timestamps with Activity Level of 0</li>
</ol>

```python
def zero_count(activities):
    zero_events = [1 for x in activities if x == 0]
    return sum(zero_events)/len(activities)
```

```python
umcontrol = [np.mean(x['activity']) for x in limited_control_df]
normalised_mean_control = [(x - min(umcontrol))/(max(umcontrol)-min(umcontrol)) for x in umcontrol]
                                           
umcondition = [np.mean(x['activity']) for x in limited_condition_df]
normalised_mean_condition = [(x - min(umcondition))/(max(umcondition)-min(umcondition)) for x in umcondition]


uscontrol = [np.std(x['activity']) for x in limited_control_df]
normalised_std_control = [(x - min(uscontrol))/(max(uscontrol)-min(uscontrol)) for x in uscontrol]

uscondition = [np.std(x['activity']) for x in limited_condition_df]
normalised_std_condition = [(x - min(uscondition))/(max(uscondition)-min(uscondition)) for x in uscondition]


normalised_zero_control = [zero_count(x['activity']) for x in limited_control_df]
normalised_zero_condition = [zero_count(x['activity']) for x in limited_condition_df]

```

```python
concat_mean = normalised_mean_condition + normalised_mean_control
concat_std = normalised_std_condition + normalised_std_control
concat_zero = normalised_zero_condition + normalised_zero_control
label = df_scores['group']

```

```python
feature_labels = ['mean', 'std', 'zero', 'label']
feature_df = pd.DataFrame(data={'mean': concat_mean, 'std': concat_std, 'zero': concat_zero, 'label': label}) 


feature_df
```

## SVM Classifier

```python
x1 = feature_df['mean']
x2 = feature_df['std']
x3 = feature_df['zero']

X = np.hstack((np.asarray(x1), np.asarray(x2), np.asarray(x3)))
X.reshape(3,55)
y = feature_df['label']
y = np.asarray(y).reshape(1,55)
X = X.reshape(3,55)
X = X.transpose()
y = y.transpose()
y = y.reshape(55,)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

print(X_train.shape, y_train.shape)
```

```python
clf = svm.SVC(kernel='linear', C=1.8).fit(X_train, y_train)
print(clf.score(X_test, y_test))

scores = cross_val_score(clf, X, y, cv=10, scoring='f1_weighted')
print(scores)
print("F1 score weighted: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))


```

```python
from sklearn.metrics import matthews_corrcoef

clf = svm.LinearSVC().fit(X_train, y_train)
y_score = clf.decision_function(X_test)
print(clf.score(X_test, y_test))
y_pred = clf.predict(X_test)
matthews_corrcoef(y_test, y_pred)
```

```python
from sklearn.metrics import average_precision_score
average_precision = average_precision_score(y_test, y_score)

print('Average precision-recall score: {0:0.2f}'.format(average_precision))
```

```python
clf = LogisticRegression(random_state=0).fit(X_train, y_train)
print(clf.score(X_test, y_test))
```

```python
gnb = GaussianNB()
clf = gnb.fit(X_train, y_train)
print(clf.score(X_test, y_test))
```
